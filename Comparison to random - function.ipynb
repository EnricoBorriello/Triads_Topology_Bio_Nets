```python
def triad_significance_profile(adjacency_matrix, ensemble_size, edge_rand_iterations):
# assumptions: no self-loops, all weights = 1

#IT = 

    ensemble = []
    
    p = tu.triad_census(matrix)

    for _ in range(ensemble_size):
        random_matrix = tu.randomize(matrix,edge_rand_iterations)
        t = list(tu.triad_census(random_matrix))
        ensemble.append(t)

        m = np.mean(ensemble,axis = 0)
        s = np.std(ensemble,axis = 0)
    
    return (p-m)/s
```
